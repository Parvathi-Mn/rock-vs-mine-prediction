```python
import pandas as pd
import numpy as np
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
sonar_data = pd.read_csv('/content/sonar data.csv',header=None)
sonar_data.head()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()  # here there are 60 coloumns, and taking the mean of all the coloumns here
```

```python
#separating the data and label, storing all the 60th column values in y
x = sonar_data.drop(columns=60,axis=1)
y = sonar_data[60]
```

```python
print(x)
print(y)
```

training and testing the data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
print(x.shape,x_train.shape,x_test.shape)
```

model training -- logistic regression model

```python
model = LogisticRegression()
```

```python
#to train the logistic regresion model with training data
model.fit(x_train,y_train)
```

```python
#evaluating the model, accuracy
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print('accuracy on training data:',training_data_accuracy)
```

```python
#evaluating the accuracy in test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
print('accuracy on test data:',test_data_accuracy)
```
